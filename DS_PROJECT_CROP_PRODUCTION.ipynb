**Finding Outliers**

```python
Q1 = df['Area'].quantile(0.25)
Q3 = df['Area'].quantile(0.75)
Q1,Q3
```

```python
IQR = Q3 - Q1
```

```python
lower_fence = Q1 -1.5 * IQR
higher_fence = Q3 + 1.5 * IQR
lower_fence,higher_fence
```

```python
area_outliers = df[(df['Area'] < lower_fence) | (df['Area'] > higher_fence)]
```

```python
# finding the number for outliers that are prosent in each crops based on the area
area_outliers['Crop'].value_counts()
```

Dividing  the states to different zones

```python
north_india = ['Jammu and Kashmir','Punjab','Himachal Pradesh','Haryana','Uttarakhand','Uttar Prdesh','Chandigarh']
east_india = ['Bihar','Odisha','Jharkhand','West Bengal']
south_india = ['Andhra Pradesh','Karnataka','Kerala','Tamil Nadu','Telungana']
west_india = ['Rajasthan','Gujarat','Goa','Maharashtra']
central_india = ['Madhya Pradesh','Chhattisgarh']
north_east_india = ['Assam','Sikkim','Nagaland','Meghalaya','Manipur','Mizoram','Tripura','Arunachal Pradesh']
ut_india = ['Andaman and Nicobar Islands','Dadra and Nagar Haveli','Puducherry']
```

```python
# creating a list for zones
zone = [
    'North India' if state in north_india else
    'East India' if state in east_india else
    'South India' if state in south_india else
    'West India' if state in west_india else
    'Central India' if state in central_india else
    'North-East India' if state in north_east_india else
    'Union Territory'
    for state in df['State_Name']
]
```

```python
# adding zone to dataset

df['Zone'] = zone
df
```

```python
# creating category of crops

# Define the crop categories dictionary
crop_categories = {
    'Cereal': ['Rice','Maize','Wheat','Barley','Varagu','Other Cereals & Millets','Ragi','Small millets','Bajra','Jowar', 'Paddy','Total foodgrain','Jobster'],
    'Pulses': ['Moong','Urad','Arhar/Tur','Peas & beans','Masoor', 'Other Kharif pulses','other misc. pulses','Ricebean (nagadal)', 'Rajmash Kholar','Lentil','Samai','Blackgram','Korra','Cowpea(Lobia)', 'Other  Rabi pulses','Other Kharif pulses','Peas & beans (Pulses)','Pulses total','Gram'],
    'Fruits': ['Peach','Apple','Litchi','Pear','Plums','Ber','Sapota','Lemon','Pome Granet', 'Other Citrus Fruit','Water Melon','Jack Fruit','Grapes','Pineapple','Orange', 'Pome Fruit','Citrus Fruit','Other Fresh Fruits','Mango','Papaya','Coconut','Banana'],
    'Beans': ['Bean','Lab-Lab','Moth','Guar seed','Soyabean','Horse-gram'],
    'Vegetables': ['Turnip','Peas','Beet Root','Carrot','Yam','Ribed Guard','Ash Gourd ','Pump Kin','Redish','Snak Guard','Bottle Gourd', 'Bitter Gourd','Cucumber','Drum Stick','Cauliflower','Beans & Mutter(Vegetable)','Cabbage', 'Bhindi','Tomato','Brinjal','Khesari','Sweet potato','Potato','Onion','Tapioca','Colocosia'],
    'Spices': ['Perilla','Ginger','Cardamom','Black pepper','Dry ginger','Garlic','Coriander','Turmeric','Dry chillies','Cond-spcs other'],
    'Fibres': ['other fibres','Kapas','Jute & mesta','Jute','Mesta','Cotton(lint)','Sannhamp'],
    'Nuts': ['Arcanut (Processed)','Atcanut (Raw)','Cashewnut Processed','Cashewnut Raw','Cashewnut','Arecanut','Groundnut'],
    'Oilseeds': ['other oilseeds','Safflower','Niger seed','Castor seed','Linseed','Sunflower','Rapeseed &Mustard','Sesamum','Oilseeds total'],
    'Commercial': ['Tobacco','Coffee','Tea','Sugarcane','Rubber']
}

# Create a reverse mapping dictionary
reverse_crop_categories = {crop: category for category, crops in crop_categories.items() for crop in crops}

# Map the crop column to their categories
df['cat_crop'] = df['Crop'].map(reverse_crop_categories)

# Display the DataFrame
df

# df.to_csv('C:\Users\H P\Desktop\Ahammed suhail\ML Datasets\crop_preprocessed_data.csv', index=False)
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='cat_crop', data=df, palette='viridis', order=df['cat_crop'].value_counts().index)
plt.xlabel('Crop Category')
plt.ylabel('Count')
plt.title('Distribution of Crops by Category')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='cat_crop', y='Production',data=df, palette='viridis')
plt.xlabel('Crop Category')
plt.ylabel('Production')
plt.yscale('log')
plt.title('Production of crops based on category')
plt.show()
```

**Top 5 crops by production**

```python

top_crops = df.groupby('Crop')['Production'].sum().nlargest(5).reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(data=top_crops, x='Crop', y='Production',palette='viridis')
plt.yscale('log')
plt.title("Overall Crops vs Production")
plt.show()
```

**Season Wise Production**

```python

top_seasons = df.groupby('Season')['Production'].sum().reset_index().sort_values(by='Production', ascending=False)
plt.figure(figsize=(15, 8))
sns.barplot(data=top_seasons, x='Season', y='Production',palette='viridis')
plt.yscale('log')
plt.title('Seasonal Crop vs Production')
plt.show()
```

```python
# production of category crops in each state
plt.figure(figsize=(10, 10))
sns.countplot(data=df, x='State_Name', hue='cat_crop', palette='magma')
plt.xticks(rotation=80)
plt.title("Category of Crop vs State")
plt.xlabel("State")
plt.ylabel("Count")
plt.legend(title='Category of Crop', loc='upper left')
plt.show()
```

```python
bar_data = df.groupby(['State_Name', 'cat_crop']).size().unstack(fill_value=0)

bar_data.plot(kind='bar', stacked=True, figsize=(12,6),colormap ='magma')
plt.xticks(rotation=80)
plt.title("Stacked Bar Plot of Crop Categories by State")
plt.xlabel("State")
plt.ylabel("Count")
plt.legend(title='Category of Crop', loc='upper left')
plt.show()
```

**Pie chart showing top 5 crop categories by count**

```python
df1=df["cat_crop"].value_counts().head(5)
df1.plot(radius=2,kind="pie",autopct="%1.1f%%",colormap='Pastel1')
plt.tick_params(labelsize=0)
```

**Top 5 categories of crop by production**

```python
# Sum production values for each crop category
category_production = df.groupby('cat_crop')['Production'].sum()

# Get the top 5 crop categories by production
top5_category_production = category_production.nlargest(5)

plt.figure(figsize=(4,4))
top5_category_production.plot(kind="pie", radius=2, autopct="%1.1f%%", colormap='Pastel1')
plt.tick_params(labelsize=0)
plt.show()
```
